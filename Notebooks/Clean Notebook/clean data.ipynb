```python
# DEALING WITH NULL VALUES

    # Building_area and ratio_green_area --> will drop it this one row

df = df[(df['building_area'].notna()==True) & (df['ratio_green_area'].notna()==True)]

    #construction_year

df['construction_year'].value_counts() # we have to see further the values that have 0.0 because maby they are meaningless

median_year = df['construction_year'].median()

df['construction_year'] = df['construction_year'].fillna(median_year)
df['construction_year'].isna().sum()

    #roof_height
    
mean_heights = df['roof_height'].mean()
df['roof_height'] = df['roof_height'].fillna(mean_heights)

    #ground_elev 

df = df[(df['ground_elev'].notna()==True)]

    #address
df = df[(df['address'].notna()==True)]

df.info()
```

```python
# ROUNDING VALUES TO BECOME EASY AND FAST TO READ

df['green_roof_area'] = df['green_roof_area'].apply(lambda x: int(x))
df['building_area'] = df['building_area'].apply(lambda x: int(x))
df['construction_year'] = df['construction_year'].apply(lambda x: int(x))
df['roof_height'] = df['roof_height'].apply(lambda x: int(x))
df['ground_elev'] = df['ground_elev'].apply(lambda x: int(x))

df['ratio_green_area'] = df['ratio_green_area'].apply(lambda x: round(x,2))

df
```

```python
#CONSTRUCTION_YEAR

df['construction_year'].value_counts().head(30)
df['construction_year'].nunique() # very dispersed values, year can be looked as a categorical variable, so it would be good to make intervals
df.describe().T 

#  we'll replace values with 0's since this values doesn's have any meaning.
    
#but first we'll try to split the years into intervals:

    #1) a good scale for this coulb be to split the years in intervals of 30 years #since the buildings within this period dont's change to much in techincal level
    #2 )  however for the buildings before 1900, the periods will be splitted in 50 years.
    

#newest_year = 2016

df[df['construction_year'] != 0].min()

#oldest_year = 1800 

#Periods to split years:

df[(df['construction_year'] >= 1800) & (df['construction_year'] < 1850)] # 2 buildings
df[(df['construction_year'] >= 1850) & (df['construction_year'] < 1900)] # 30 buildings
df[(df['construction_year'] >= 1900) & (df['construction_year'] < 1930)] # 187 buildings
df[(df['construction_year'] >= 1930) & (df['construction_year'] < 1960)]#129 buildings
df[(df['construction_year'] >= 1960) & (df['construction_year'] < 1990)]#98 buildings
df[(df['construction_year'] >= 1990)] #248 buildings
                                      #Total = 733 buildings ( with 0's, ok!)

def clean_year(x):
    if 1800 <= x < 1850:
        return '1800'
    elif 1850 <= x < 1900:
        return '1850'
    elif 1900 <= x < 1930:
        return '1900'
    elif 1930 <= x < 1960:
        return '1930'
    elif 1960 <= x < 1990:
        return '1960'
    elif 1990 <= x :
        return '1990'
    else:
        return x
    

df['construction_year'] = df['construction_year'].apply(clean_year)
```

```python
round(df['construction_year'].value_counts()*39/(df.shape[0]),0), # to know hot to fill the 0's in a propotional way
```

```python
#Replacing 0's in 'column_year'

df = df.reset_index()
rows_zeros = list(df[df['construction_year'] == 0].index)
df.iloc[rows_zeros[0:13],[5]] = df.iloc[rows_zeros[0:13],[5]].apply(lambda x: x.replace(0, '1990'))
df.iloc[rows_zeros[13:23],[5]] = df.iloc[rows_zeros[13:23],[5]].apply(lambda x: x.replace(0, '1900'))
df.iloc[rows_zeros[23:30],[5]] = df.iloc[rows_zeros[23:30],[5]].apply(lambda x: x.replace(0, '1930'))
df.iloc[rows_zeros[30:35],[5]] = df.iloc[rows_zeros[30:35],[5]].apply(lambda x: x.replace(0, '1960'))
df.iloc[rows_zeros[35:38],[5]] = df.iloc[rows_zeros[35:38],[5]].apply(lambda x: x.replace(0, '1850'))
df.iloc[rows_zeros[38:39],[5]] = df.iloc[rows_zeros[38:39],[5]].apply(lambda x: x.replace(0, '1800'))
```

```python
# after some research, we find that the column 'zonedist1' that was dropped,
    # is refering to the type of areas of the city(commercial, industrial, residential,public buildings)

#so we want to keep this column and treat that

df2 = pd.read_csv('../../Data/raw_data/GreenRoofData2016_20180917.csv')

df2['zonedist1'] = df2['zonedist1'].apply(str)

def clean_type_zone(x):
    if x[0] =='C':
        return 'commercial'
    elif x[0] =='M':
        return 'industrial'
    elif x[0] =='R':
        return 'residential'
    elif x[0] =='B':
        return 'residential'
    elif x[0] =='P':
        return 'public'
    else:
        return x
    
df2['zonedist1'] = df2['zonedist1'].apply(clean_type_zone)

df2['zonedist1'].value_counts()
```
